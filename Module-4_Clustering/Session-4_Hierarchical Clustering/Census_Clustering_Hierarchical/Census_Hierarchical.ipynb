```python
import pandas as pd

df = pd.read_excel("/content/DS_India_Census_Cleaned.xlsx")
df.head()
```

```python
new_df = df.iloc[:,4:]
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
new_df_sc = scaler.fit_transform(new_df)
new_df_sc = pd.DataFrame(new_df_sc, columns = new_df.columns)
new_df_sc.head()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
import matplotlib.pyplot as plt
```

```python
merged = linkage(new_df_sc, method = 'single')
dendrogram(merged)
plt.show()
```

```python
merged = linkage(new_df_sc, method = 'complete')
dendrogram(merged)
plt.show()
```

```python
merged = linkage(new_df_sc, method = 'average')
dendrogram(merged)
plt.show()
```
