```python
import pandas as pd

df = pd.read_csv("/content/DS_Cricket.csv")
df.head()
```

```python
new_df = df[['Ave', 'SR']]
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
new_df_sc = scaler.fit_transform(new_df)
new_df_sc = pd.DataFrame(new_df_sc, columns = new_df.columns)
```

```python
new_df_sc.head()
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 4, random_state=100)
model.fit(new_df_sc)
```

```python
model.labels_
```

```python
df['Label'] = model.labels_
```

```python
df.head()
```

```python
# (high SR, high Ave) - A, (low SR, low Ave) - B, (High SR, Low Ave) - C, (Low SR, High Ave) - D

df[df.Label == 0]
```

# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
import matplotlib.pyplot as plt
```

```python
merged = linkage(new_df_sc, method = 'complete')
dendrogram(merged)
plt.figure(figsize=(20,10))
plt.show()
```

```python
labels = cut_tree(merged,n_clusters=4)
labels.shape
```

```python
labels.reshape(-1,)
labels.shape
```

```python
df['Hierar_Label'] = labels
```
